```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from keras.datasets import fashion_mnist
```

```python
(x_train,y_train1),(x_test,y_test1) = fashion_mnist.load_data()
```

```python
(x_train,y_train1),(x_test,y_test1) = (x_train/255,y_train1),(x_test/255,y_test1)
```

```python
fig,axes = plt.subplots(15,15,figsize=(20,20))
axes = axes.ravel()
for i in range(225):
    axes[i].imshow(x_train[i,1:])
    axes[i].set_title(y_train1[i],fontsize=8)
    axes[i].axis("off")
                
plt.subplots_adjust(hspace=0.4)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
y_train = encoder.fit_transform(y_train1.reshape(-1,1)).toarray()
y_test = encoder.fit_transform(y_test1.reshape(-1,1)).toarray()
```

```python
y_train
```

```python
plt.imshow(x_train[0],cmap="gray")
```

```python
x_train = x_train.reshape(x_train.shape[0],28,28,1)
x_test = x_test.reshape(x_test.shape[0],28,28,1)
```

```python
from keras.layers import Conv2D,Dense,Flatten,Dropout,MaxPool2D
from keras.models import Sequential
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
model = Sequential()
model.add(Conv2D(128,(3,3),input_shape=(28,28,1),activation="relu"))
model.add(MaxPool2D((2,2)))
model.add(Conv2D(64,(3,3),activation="relu"))
model.add(MaxPool2D((2,2)))
model.add(Dropout(.5))
model.add(Conv2D(32,(3,3),activation="relu"))
model.add(MaxPool2D((2,2)))
model.add(Flatten())
model.add(Dense(output_dim=1000,activation="relu"))
model.add(Dropout(.5))
model.add(Dense(output_dim=32,activation="relu"))
model.add(Dropout(.3))
model.add(Dense(output_dim=10,activation="softmax"))
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.summary()
```

```python
history = model.fit(x_train,y_train,batch_size=10,epochs=10,verbose=1,validation_data=(x_test,y_test))
```

```python
history.history
```

```python
plt.plot(history.history["acc"])
plt.plot(history.history["val_acc"])
plt.title("model accuracy")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend(["train","test"])
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("loss")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend(["train","test"])
```

```python
model.save("fashion_model_weights3.h5")
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
prediction = model.predict_classes(x_test)
```

```python
classification_report(prediction,y_test1)
```

```python
model.evaluate(x_test,y_test)
```
